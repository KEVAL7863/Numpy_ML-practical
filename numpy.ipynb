Stacking Array ----> Means two array ko joad do (stack kar do)

```python
#   vartical stacking ----->  usme hamesha vector ka use hota hai not uding matrix
a = np.array([1,2,3,4,5,6])
b = np.array([7,8,9,10,11,12])
```

```python
np.vstack((a,b))
```

```python
# varical stack ko call karo or a and b ko tuple ke form  me bhej do

np.hstack((a,b)) # horizotnal means dono aaray ko joad dena

```

```python
np.column_stack((a,b)) # hstack ye 1D array hai to uska muje 2D array karne ke liye colum_stack ka use kare
```

splitting Array


```python
c = np.arange(1,17).reshape(4,4)
c
```

```python
# split ko horizontal or vertical me split kar sakte hai
# muje horizotal ko split ke liye bich se cutout ho jayega and muje 2 array create ho jayega or do bhag me divide ho jayega 

np.hsplit(c,2) #ya np.hspli(c,4),(c,1) use kar sakte
```

```python
# muje verical ko split ke liye bich se cutout ho jayega and muje 2 array create ho jayega or do bhag me divide ho jayega 

a =np.vsplit(c,4) #ya np.hspli(c,4),(c,1) use kar sakte
```

```python
for i in a:
    print(i)
```
